1. What is the difference between a function and a method in Python?

-> the key difference between a function and a method lies in their association with objects:

Function
a. A function is a standalone block of code that performs a specific task.

b. It is defined using the def keyword and can be called anywhere in the program.

c. It is not tied to any object.

Example: in code given below:-

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Abhay"))
```

Method
a. A method is a function that belongs to an object.

b. It is defined inside a class and operates on the data of the instance.

c. Methods are invoked using the dot notation.

d. The first parameter of an instance method is usually self, which refers to the instance.

Example: code given below:-

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

person = Person("Abhay")
print(person.greet())
```

->In Python, parameters and arguments are related but distinct concepts when working with functions:

Parameters → Variables listed in a function definition (placeholders).

Arguments → Actual values passed to a function when calling it.

3. What are the different ways to define and call a function in Python?

-> a. Regular Function Definition and Call
Definition: Using the def keyword.
Call: Using the function name with parentheses.
example:- code given below:-

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Abhay"))
```

b. Function with Default Arguments
Definition: Provides default values for parameters.
Call: Can be called with or without arguments.

```python
def welcome(name="Abhay"):
    return f"Welcome, {name}!"

print(welcome())
print(welcome("Raj"))
```

c. Function with *args (Variable-Length Positional Arguments)
Definition: Allows multiple arguments as a tuple.
Call: Pass any number of values.

```python
def hobbies(name, *hobbies):
    print(f"{name}'s hobbies are: {', '.join(hobbies)}")

hobbies("Abhay", "Reading", "Gaming", "Cycling")
```

d. Lambda (Anonymous) Function
Definition: Single-line function using lambda.

Call: Call like a normal function.

```python
double = lambda x: x * 2
print(double(5))
```

5.  What are iterators in Python and how do they differ from iterables?

-> An iterable is any object that can be looped over (e.g., lists, tuples, strings, dictionaries, sets).
It contains data and has a built-in method __iter__() that returns an iterator.
Example:-

```python
names = ["Abhay", "Raj", "Neha"]  # list -> iterable

for name in names:
    print(name)
```

An iterator is an object that produces data one element at a time using __next__().
It remembers its current position in the sequence.
example:-

```python
names = ["Abhay", "Raj", "Neha"]
names_iterator = iter(names)

print(next(names_iterator))
print(next(names_iterator))
print(next(names_iterator))
```

6.  Explain the concept of generators in Python and how they are defined.

-> A generator is a function that produces a sequence of values one at a time using the yield keyword, instead of returning all values at once like a normal function. They generate values on the fly, which makes them memory-efficient and suitable for working with large datasets. Example:-

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
gen = count_up_to(3)
print(next(gen))
print(next(gen))
print(next(gen))
```

8.  What is a lambda function in Python and when is it typically used?

-> A lambda function in Python is a small, anonymous function that can have any number of arguments but only one expression. It is typically used when a short, throwaway function is needed without formally defining a function using def.

```python
square_lambda = lambda x: x ** 2
print(square_lambda(5))
```

9.  Explain the purpose and usage of the `map()` function in Python.

->The map() function in Python is used to apply a function to each element in an iterable (like a list or tuple) and returns a new iterable (a map object). This is useful when you need to transform data efficiently without using explicit loops. Example:-

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

10. What is the difference between map(), reduce(), and filter() functions in Python?

-> ython provides three powerful higher-order functions for functional programming:

 map() : Transforms elements of an iterable.

 filter() : Filters elements based on a condition.

 reduce() : Reduces elements to a single value.

```python
numbers0 = [1, 2, 3, 4, 5]

# Using map() to square each element
squared = list(map(lambda x: x ** 2, numbers0))

print(squared)


numbers1 = [1, 2, 3, 4, 5, 6, 7, 8]

# Using filter() to keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers1))

print(evens)

from functools import reduce

numbers2 = [1, 2, 3, 4, 5]

# Using reduce() to find the product of all elements
product = reduce(lambda x, y: x * y, numbers2)

print(product)

```

```python
# Question 3:   Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

list1 = list(map(int, input("Enter a list of numbers seperated by space:- ").split()))
ans_list = []

def square_list(list):
    for i in list:
        ans_list.append(i**2)
    return ans_list

print(square_list(list1))
```

```python
# Question 4:   Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
prime_numbers = [num for num in range(1, 201) if is_prime(num)]

print("Prime numbers from 1 to 200:", prime_numbers)

```

```python
# Question 5 : Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci_no(n):
  a, b = 0, 1
  for _ in range(n):
      yield a
      a, b = b, a+b
print(fibonacci_no(10))
print(list(fibonacci_no(10)))
```

```python
# Question 6:  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for exponent in range(n + 1):
        yield exponent ** 2
print(list(powers_of_two(5)))
```

```python
# Question 7:  Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
# Question 8:  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# List of tuples (name, age)
data = [("Abhay", 25), ("Rohan", 20), ("Sneha", 30), ("Meera", 22)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
# Question 9:  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

celsius_temps = [0, 20, 37, 100, -10]
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))
print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

```python
# Question 10 : Create a Python program that uses `filter()` to remove all the vowels from a given string.

my_str = input("Enter a string:- ")
ans_str = list(filter(lambda x: x.lower() not in 'aeiou', my_str))
print("".join(ans_str))
```

```python
''' Question 11 :  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
 Write a Python program using lambda and map'''

def process_orders(order_list):
    result = list(map(
        lambda order: (
            order[0],
            order[3] * order[2] + (10 if order[3] * order[2] < 100 else 0)
        ),
        order_list
    ))

    return result

orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

result = process_orders(orders)

print("Order Number | Total Value (€)")
print("-" * 30)
for order_num, total in result:
    print(f"{order_num:<12} | {total:.2f}")
```
